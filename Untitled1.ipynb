```python
#최종완성품
#

n=int(input("숫자를 입력하시오 :"))
a=[n]
b=len(a)
for i in range(100):
  c=int(input("계속 입력을 원하시면, 자연수 한 개를 입력하시고, 아니라면 0을 입력하세요 : "))
  
  if c!=0:
    a.append(c) 
    for k in range(len(a)):
      for j in range(k+1,len(a)):
        if a[k]<a[j]:
          a[k],a[j]=a[j],a[k]    

  else:
    print("당신이 입력한 숫자는 큰 것부터 순서대로 말한다면 아래와 같습니다.",a)
    break
if len(a)>=100:
  print("이제그만하세요. 벌써 100개를 입력했어요")
```
